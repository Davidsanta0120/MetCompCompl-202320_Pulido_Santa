Punto 2

Tomando las siguientes constantes, M=1, R=8,31446

```python
M=1
R=8.31446
Probv= lambda v,T: 4*np.pi*(M/(2*np.pi*R*T))**(3/2)*v**2*np.exp(-(M*v**2)/(2*R*T))

V=np.linspace(0,200,101)

T=np.linspace(100,300,5)



for i in T:
    y=Probv(V,i)
    plt.plot(V,y,label="T="+str(int(i)))
plt.legend()
plt.show()
```

A partir de las gráficas se puede evidenciar que, al aumentar la temperatura, la probabilidad de la velocidad máxima disminuye, a su vez que el valor donde se presenta esta velocidad máxima aumenta.


Punto 3

Haciendo la misma sustitucion que en el punto 1, obtenemos $v_{avg}=\sqrt{\frac{8RT}{M\pi}} \int_{0}^{\infty} ue^{-u} \, du$


```python
M=1
R=8.31446
Root,Weight = np.polynomial.laguerre.laggauss(7)

Vprom=lambda T:np.sqrt((8*R*T)/(M*np.pi))*np.sum(Weight*Root)
T2=np.linspace(100,5000,10)
Vavg=Vprom(T2)
plt.scatter(T2,Vavg)

```

Como $\int_{0}^{\infty} ue^{-u} \, du =1$ por lo tanto, $v_{avg}=\sqrt{\frac{8RT}{M\pi}}$
